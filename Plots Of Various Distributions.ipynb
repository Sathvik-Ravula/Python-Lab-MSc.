***Binomial Distribution***

```python
import numpy as np
import math
import matplotlib.pyplot as plt


def binom(n,p,x):
    return math.comb(n,x)*p**x*(1-p)**(n-x)

n = int(input("Enter No. Of Trails :"))
p = float(input("Enter P(success) :"))

X = [i for i in range(n+1)]
y = []
for x in X:
    y.append(binom(n,p,x))
    
plt.title(f'Plot of Binomial Random Variable from {X[0]} to {X[-1]}')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)   
plt.plot(X,y,'g-')
plt.bar(X,y)
plt.show()
```

***Normal Distribution***

```python
import numpy as np
import math
import matplotlib.pyplot as plt


def normal(m,s,x):
    return 1/(np.sqrt(2*np.pi*s**2)) * np.exp(-1/2 * (x-m/s**2)**2)

m = float(input("Enter Mean Value :"))
s = float(input("Enter Sigma Value :"))

X = np.linspace(m-4*s,m+4*s,500)
y = []
for x in X:
    y.append(normal(m,s,x))
    
plt.title(f'Plot of Normal Random Variable from {X[0]} to {X[-1]}')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)  
plt.plot(X,y,'b-')
plt.show()
```

***Exponential Distribution***

```python
import numpy as np
import math
import matplotlib.pyplot as plt


def exponential(l,x):
    return l*np.exp(-l*x)

l = float(input("Enter lambda value :"))

X = np.linspace(0,20,500)
y = []
for x in X:
    y.append(exponential(l,x))
    
plt.title(f'Plot of Exponential Random Variable from {X[0]} to {X[-1]}')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)   
plt.plot(X,y,'r')
plt.show()
```

***Geometric Distribution***

```python
import numpy as np
import math
import matplotlib.pyplot as plt


def geometric(n,p,x):
    return p*(1-p)**(x-1)

p = float(input("Enter P(success) :"))
n = int(input("Enter N Value :"))

X = [i for i in range(n+1)]
y = []
for x in X:
    y.append(geometric(n,p,x))
    
plt.title(f'Plot of Geometric Random Variable from {X[0]} to {X[-1]}')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)   
plt.plot(X,y,'g-')

plt.show()
```

***Gamma Distribution***

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.integrate as integrate

def gamma_integral(a,x):
    return x**(a-1) * np.exp(-x)

def gamma_rv(a,l,x):
    return ((l**a)/gamma_integral(a,x))*x**(a-1) * np.exp(-l*x)

l = float(input("Enter lambda value :"))
a = float(input("Enter alpha value :"))

X = np.linspace(0,10,500)
y = []
for x in X:
    y.append(gamma_rv(a,l,x))
    
plt.title(f'Plot of Gamma Random Variable from {X[0]} to {X[-1]}')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)   
plt.plot(X,y,'r')
plt.show()
```

***Poisson Distribution***

```python
import numpy as np
import math
import matplotlib.pyplot as plt


def poisson(l,x):
    return ((l**x)*np.exp(-l))/math.factorial(x)

l = int(input("Enter Lambda Value :"))

X = [i for i in range(21)]
y = []
for x in X:
    y.append(poisson(l,x))
    
plt.title(f'Plot of Poisson Random Variable from {X[0]} to {X[-1]}')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)   
plt.plot(X,y,'r-')

plt.show()
```
